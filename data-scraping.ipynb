```python
import yfinance as yf
import pandas as pd

# Stock tickers from the paper
tickers = {
    'NKE': 'Nike',
    'GS': 'Goldman Sachs',
    'JNJ': 'Johnson and Johnson',
    'PFE': 'Pfizer',
    'JPM': 'JP Morgan Chase'
}

all_data = {}

for ticker, name in tickers.items():
    data = yf.download(ticker, start="2009-04-05", end="2019-04-05")
    data.columns = data.columns.droplevel(1)
    data.columns.name = None
    data.reset_index(inplace=True)

    # Create new variables as per the paper
    data['H-L'] = data['High'] - data['Low']
    data['O-C'] = data['Close'] - data['Open']
    data['7_DAY_MA'] = data['Close'].rolling(window=7).mean()
    data['14_DAY_MA'] = data['Close'].rolling(window=14).mean()
    data['21_DAY_MA'] = data['Close'].rolling(window=21).mean()
    data['7_DAY_STD'] = data['Close'].rolling(window=7).std()
    
    all_data[ticker] = pd.DataFrame(data)

# Display sample from last ticker
all_data['JPM'].tail()
```

```python
for ticker, data in all_data.items():
    data.to_csv(f'./data/{ticker}_stock_data.csv', index=False)
```
